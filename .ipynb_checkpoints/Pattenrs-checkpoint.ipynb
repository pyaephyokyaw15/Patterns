# Full Number Pyramid

```python
rows = int(input("Enter number of rows: "))

k = 0
count=0
count1=0

for i in range(1, rows+1):
    for space in range(1, (rows-i)+1):
        print("  ", end="")
        count+=1
    
    while k!=((2*i)-1):
        if count<=rows-1:
            print(i+k, end=" ")
            count+=1
        else:
            count1+=1
            print(i+k-(2*count1), end=" ")
        k += 1
    
    count1 = count = k = 0
    print()
```

```python
rows = int(input("Enter number of rows: "))

for i in range(1, rows + 1):
    # print spaces
    print('  ' * (rows-i), end='')
    
    # print numbers
    no_columns = 2*i - 1
    for j in range(1, no_columns + 1):
        print(no_columns - abs(i-j), end=' ') 

    print()
```

# Character Half Pyramid

```python
code = 65
for i in range(5):
    for j in range(i + 1):
        print(chr(code + j), end=' ')
    print()
```

```python
rows = int(input('Enter rows: '))

numbers = [[] for i in range(rows)]


for i in range(rows):
    # spaces
    print(' ' * (rows - (i+1)), end='')

    # numbers
    for j in range(i+1):
        if i > j:
            # serve two options.
                # 1. greater than 0.(check at least two row)
                # 2. solve out of range for last column 
            if j == 0:
                numbers[i].append(1)
                print(1, end=' ')
            else:
                number = numbers[i-1][j-1] + numbers[i-1][j]
                numbers[i].append(number)
                print(number, end=' ')
            
        elif i == j: # also valid for first row(only one element)
            numbers[i].append(1)
            print(1, end=' ')

    print()

            
      
 
print(numbers)
```

# Floyd Triangle

```python
number= 1
for i in range(1, 6):
    for j in range(1, i+1):
        print(number, end=' ')
        number += 1
    print()
```
